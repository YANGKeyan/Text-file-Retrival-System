```python
import os

# 示例控制词汇列表
control_vocabulary = ["covid","covid.19", "omicron", "prophecy", "not Omicron"]

# 初始化文档索引
document_index = {}

# 遍历预处理文件夹中的每个txt文件
preprocessed_dir = "preprocessed"
for filename in os.listdir(preprocessed_dir):
    if filename.endswith(".txt"):
        with open(os.path.join(preprocessed_dir, filename), "r", encoding="utf-8") as file:
            text = file.read()

        # 在文本中查找包含的控制词汇
        included_terms = []
        for term in control_vocabulary:
            if term in text:
                included_terms.append(term)

        # 将文档及其包含的控制词汇添加到文档索引中
        document_index[filename] = included_terms

# 打印文档索引
print("Document Index:")
for document, terms in document_index.items():
    print(document, ":", terms)
```

```python
import os

# 示例控制词汇列表
control_vocabulary = ["covid.19", "covid-19 and Omicron", "covid-19 and Prophecy", "Covid-19 not Omicron"]

# 示例文档索引（假设已经建立）
document_index = {
    "document1.txt": ["covid.19", "covid-19 and Omicron"],
    "document2.txt": ["covid.19", "Covid-19 not Omicron"],
    # 其他文档及其包含的控制词汇
}

# 控制词汇搜索功能实现
def controlled_vocabulary_search(query):
    relevant_documents = []
    for term in query.split():
        if term in control_vocabulary:
            relevant_documents.extend(document_index.get(term, []))
    return relevant_documents

# 示例查询
query = "covid.19"
relevant_documents = controlled_vocabulary_search(query)
print("Relevant Documents:", relevant_documents)
```
